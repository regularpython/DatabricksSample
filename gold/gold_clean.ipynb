```python
# Extract
silver_df = spark.read.parquet(
    "/Volumes/dsr_catalog/medallion/silver_vol/cleaned_sales"
)

from pyspark.sql.functions import sum

# Tranform
gold_df = silver_df.groupBy("product") \
                   .agg(sum("price").alias("total_sales"))

# Load
gold_df.write.mode("overwrite").parquet(
    "/Volumes/dsr_catalog/medallion/gold_vol/aggregated_sales"
)

# display(gold_df)
```
