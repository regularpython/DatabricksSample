```python
# Extract
bronze_df = spark.read.csv(
    "/Volumes/dsr_catalog/medallion/bronze_vol/raw_sales.csv",
    header=True,
    inferSchema=True
)

# Tranformation
silver_df = bronze_df.filter(bronze_df.quantity > 0)

# Load
silver_df.write.mode("overwrite").parquet(
    "/Volumes/dsr_catalog/medallion/silver_vol/cleaned_sales"
)

# display(silver_df)
```
